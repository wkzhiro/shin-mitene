```python
from azure.search.documents import SearchClient
from azure.core.credentials import AzureKeyCredential
import os
from dotenv import load_dotenv
```

```python
# Azure AI Searchã®è¨­å®š
search_endpoint = os.getenv('COG_END_POINT')
search_credential = AzureKeyCredential(os.getenv('COG_API_KEY'))
index_name = os.getenv('INDEX_NAME')

print(search_endpoint,search_credential,index_name)
```

```python

# Azure AI Searchã‚¯ãƒ©ã‚¤ã‚¢ãƒ³ãƒˆã®åˆæœŸåŒ–
search_client = SearchClient(
    endpoint=search_endpoint,
    index_name=index_name,
    credential=search_credential
)

search_query="python"

# Azure AI Searchã§æ¤œç´¢
search_response = search_client.search(
    search_text=search_query,
    select=['chunk_id', 'chunk', 'tags', 'parent_filename', 'creation_date'],
)

#æ¤œç´¢çµæœã‚’ãƒªã‚¹ãƒˆåŒ–
results = list(search_response)

# æ¤œç´¢çµæœã‚’å‡¦ç†
for result in results:
    print(f"Chunk ID: {result['chunk_id']}")
    print(f"Chunk Content: {result['chunk']}")
    print(f"Tags: {result['tags']}")
    print(f"Parent Filename: {result['parent_filename']}")
    print(f"Creation Date: {result['creation_date']}")
```
